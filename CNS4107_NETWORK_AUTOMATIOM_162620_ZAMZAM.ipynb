'ConnectHandler' acts as a **universal dapter**.

#### Roles of 'ConnectHandler'
1. Establishes SSH connection to a network device and
2. Automatically handles the "handshake" (logging in, handles the # or > prompts.
**NB:** The moment you specify a device_type e.g. cisco_ios, 'ConnectHandler' knows how to
- specifically navigate that specific 05,
- enter configuration mode, and
- wait for the device to finish processing a command.
